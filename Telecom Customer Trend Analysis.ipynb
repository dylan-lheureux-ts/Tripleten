## Aggregate data per user


```python
# Calculate the number of calls made by each user per month. Save the result.

calls['month'] = calls['call_date'].dt.to_period('M')
calls_monthly_per_user = calls.groupby(['user_id', 'month']).agg({
    'id': 'count',        # number of calls
}).reset_index()

print(calls_monthly_per_user)
```

```python
# Calculate the amount of minutes spent by each user per month. Save the result.
calls['month'] = calls['call_date'].dt.to_period('M')
minutes_monthly_per_user = calls.groupby(['user_id', 'month']).agg({
    'duration': 'sum'     # total minutes
}).reset_index()

print(minutes_monthly_per_user)
```

```python
# Calculate the number of messages sent by each user per month. Save the result.
messages['month'] = messages['message_date'].dt.to_period('M')
messages_monthly_per_user = messages.groupby(['user_id', 'month']).agg({
    'id': 'count',        # number of messages
}).reset_index()

print(messages_monthly_per_user)
```

```python
# Merge the data for calls, minutes, messages, internet based on user_id and month
merged_data = calls_monthly_per_user.merge(messages_monthly_per_user, on=['user_id', 'month'], how='outer')
merged_data = merged_data.merge(minutes_monthly_per_user, on=['user_id', 'month'], how='outer')
merged_data = merged_data.merge(internet_monthly_mb, on=['user_id', 'month'], how='outer')
merged_data = merged_data.rename(columns = {'id_x':'num_calls', 'id_y':'num_messages'})
merged_data = merged_data.fillna(0)
print(merged_data)
```

```python

# Add the plan information
merged_data = merged_data.merge(users[['user_id', 'plan']], on='user_id', how='left')
print(merged_data)
```

```python
# Calculate the monthly revenue for each user
        
def calculate_monthly_revenue(row):
    # Base monthly fee
    if row['plan'] == 'surf':
        base_fee = 20
        # Calculate overages
        extra_minutes = max(0, row['duration'] - 500)
        extra_messages = max(0, row['num_messages'] - 50) 
        extra_gb = max(0, row['gb_used'] - 15)
        
        revenue = base_fee + (extra_minutes * 0.03) + (extra_messages * 0.03) + (extra_gb * 10)
        
    elif row['plan'] == 'ultimate':
        base_fee = 70
        # Calculate overages
        extra_minutes = max(0, row['duration'] - 3000)
        extra_messages = max(0, row['num_messages'] - 1000)
        extra_gb = max(0, row['gb_used'] - 30)
        
        revenue = base_fee + (extra_minutes * 0.01) + (extra_messages * 0.01) + (extra_gb * 7)
    
    return revenue

# Add revenue column to your DataFrame
merged_data['monthly_revenue'] = merged_data.apply(calculate_monthly_revenue, axis=1)

# Total Revenue by plan
total_revenue_by_plan = merged_data.groupby('plan')['monthly_revenue'].sum()
mean_revenue_by_plan = merged_data.groupby('plan')['monthly_revenue'].mean()
print("Total Revenue by Plan:")
print(total_revenue_by_plan)
print("Mean Revenue by Plan:")
print(mean_revenue_by_plan)
```

## Test statistical hypotheses

**Hypothesis Setup:**

• H₀ (Null Hypothesis): μ_surf = μ_ultimate (mean revenues are equal), I want to find if a difference exists.

• H₁ (Alternative Hypothesis): μ_surf ≠ μ_ultimate (mean revenues are different)

• Significance Level: α = 0.05, does not have a life-threatening risk-related industry.

• Test Choice: Welch's t-test (unequal variances). The test samples are independent of each other, and when tested, their variances are not equal. I am also running a 2-tailed test to assess for any difference.

• Assumptions: Independent samples, approximately normal distributions

```python

# Test the hypotheses
array1 = merged_data[merged_data['plan'] == 'surf']['monthly_revenue']
array2 = merged_data[merged_data['plan'] == 'ultimate']['monthly_revenue']
alpha = 0.05

results = st.ttest_ind(array1, array2, equal_var=False)

print('p-value', results.pvalue)
if results.pvalue < alpha:
    print("We Reject the Null Hypothesis")
else:
    print("We Can't Reject the Null Hypothesis")

```

What are the odds that the surf and ultimate plans will produce the same revenue? The answer to that is 3.17*10^-13% or virtually 0%. Meaning that we can statistically say that we reject the null hypothesis, and there is a statistically significant difference between the plans, and the plan that is chosen does impact the revenue.
